```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
from sympy.solvers import solve
from sympy import Symbol, pi, exp
from scipy.optimize import brentq
from matplotlib import pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_excel('HE_Data.xlsx', sheetname='All_Data')

# RMS Values for each day
VI = [(0.505, 0.0000048), (0.544, 0.0000052), (0.543, 0.0000052)]

# Show head of data
df.head()
```

```python
## Calculating Resistivity:
# Start with measurements on day 6:
days = ['6','7','8']

RA = []
RB = []
for name in list(df):
    # Get current for the day
    I = VI[days.index(str(name)[0])][1]
    
    # Determine R's
    R1243 = df.loc[(12,43,'V'), name]/I
    R3421 = df.loc[(34,21,'V'), name]/I
    R3241 = df.loc[(32,41,'V'), name]/I
    R1423 = df.loc[(14,23,'V'), name]/I

    # Calculate RA's
    RA.append(R1243 + R3421)
    RB.append(R3241 + R1423)
```

```python
# Solve for RS
RS = Symbol("RS")
def opt_fun(ra, rb, RS):
    return ((exp(-pi*ra/RS) + exp(-pi*rb/RS) - 1)**2)

RS1 = []
for ra,rb in zip(RA,RB):
    def opt_fun(RS):
        return exp(-pi*ra/RS)+ exp(-pi*rb/RS) - 1
    RS1.append(brentq(opt_fun,100000,100000000))

resis_df = pd.DataFrame(np.reshape(RS1,[1,-1]), index=['Res'], columns=list(df))
DF = df.append(resis_df)
DF = DF.T
```

```python
# Box Plot of Resistivity:
plt.boxplot(DF['Res'],0,'gD')
plt.title('Box Plot of Resistivity Measurements')
plt.xlabel('Resistivity')
plt.show()

# Resistivity versus Temperature
plt.scatter(DF[('N','N','Temp (K)')], DF['Res'])
plt.title('Resistivity Versus Temperature')
plt.xlabel('Temperature (K)')
plt.ylabel('Resistivity (Ohm)')
plt.show() 

# Resistivity versus B field
plt.scatter(DF[('N','N','B Field (kG)')], DF['Res'])
plt.title('Resistivity Versus B Field')
plt.xlabel('B Field (kG)')
plt.ylabel('Resistivity (Ohm)')
plt.show() 
```

```python
Bs = np.asarray(DF[('N','N','B Field (kG)')])
B_fields = np.unique(abs(Bs))
Ts = np.asarray(DF[('N','N','Temp (K)')])
Temps = np.unique(abs(Ts))
VH_all = []
T_all = []
B_all = []
# Loop over each entry in the dataframe
for index, row in DF.iterrows():
    T = row[('N', 'N', 'Temp (K)')]
    B = row[('N', 'N', 'B Field (kG)')]

    # If the B field is the positive version, do the analysis
    if (np.sign(B)==1 and B!=0):
        # Get the associated negative B field entry
        dataPOS = row
        dataNEG = DF[(DF[('N', 'N', 'B Field (kG)')]==(-B)) & (DF[('N', 'N', 'Temp (K)')]==T)].iloc[0,:]
        
        # Calculate values
        V31P = dataPOS[((31, 42, 'V'))]
        V31N = dataNEG[((31, 42, 'V'))]
        V24P = dataPOS[((24, 13, 'V'))]
        V24N = dataNEG[((24, 13, 'V'))] 
        V31 = V31P - V31N
        V24 = V24P - V24N
        
        # Calculate VH
        VH = (V31 + V24)/4
        
        # Add data to the list
        T_all.append(T)
        B_all.append(B)
        VH_all.append(VH)
        
# Create a new dataframe:
d = {'Temp':T_all, 'B_Field':B_all, 'VH':VH_all}
VH_DF = pd.DataFrame(data=d)
```

```python
# Plot VH versus T at 2.5 kG
plt.scatter(VH_DF[VH_DF['B_Field']==2.5].iloc[:,1], VH_DF[VH_DF['B_Field']==2.5].iloc[:,2])
plt.title('Hall Voltage Versus Temperature at B=+-2.5 kG')
plt.xlabel('Temperature (K)')
plt.ylabel('Hall Voltage (V)')
plt.show()

# Plot VH versus B at 82.4 K
plt.scatter(VH_DF[VH_DF['Temp']==82.4].iloc[:,0], VH_DF[VH_DF['Temp']==82.4].iloc[:,2])
plt.title('Hall Voltage Versus B Field at T=82.4 K')
plt.xlabel('B Field (kG)')
plt.ylabel('Hall Voltage (V)')
plt.show()
```

```python
# Model the B field at T=82.4
Bs = VH_DF[VH_DF['Temp']==82.4].iloc[:,0].values.reshape(-1,1)
VHs = VH_DF[VH_DF['Temp']==82.4].iloc[:,2]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(Bs, VHs)

# Make predictions using the testing set
reg_out = regr.predict(Bs)

# The coefficients and scores
print('Coefficients: \n', regr.coef_)
print('Intercept: \n',regr.intercept_)
print('Linear Model: \n VH = %.4f*B + %.4f' % (regr.coef_, regr.intercept_))
print('R Squared: \n',regr.score(Bs,VHs))

# Showing Plot of Hall Voltages
plt.plot(Bs, reg_out, color='blue', linewidth=3)
plt.scatter(Bs, VHs)
plt.xlabel('Magnetic Field (kG)')
plt.ylabel('Hall Voltage (H)')
plt.title('Dependence of Hall Voltage on Magnetic Field at T=82.4 K')
plt.show()
```
